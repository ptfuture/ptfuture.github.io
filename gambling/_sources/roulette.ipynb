Assuming that the roulette is fair and each bet is iid. Let $p$ be the prability of winning, $q$ be the opposite. Let $X$ be the rewards of winning for betting one dollar, or $-1$ if loses.

## Simple Strategies

### Single number betting (Straight-up)

$X=35$ when $p=\frac{1}{37}$, or $X=-1$ when loses. 

$$
\begin{aligned}
E(X) & = 35\times p - q = \frac{35}{37} - \frac{36}{37} = - \frac{1}{37} = -0.02702702702 \\
E(X^2) & = 35^2 \times p + q = \frac{35^2}{37} + \frac{36}{37} = 34.0810810811 \\
Var(X) & = E(X^2) - E(X)^2 = 34.0810810811 - (-0.02702702702)^2 = 34.0803506209
\end{aligned}
$$

### Two numbers betting (Split)

$X=17$ when $p=\frac{2}{37}$, or $X=-1$ when loses.

$$
\begin{aligned}
E(X) & = 17 \times p - q = 17 \times \frac{2}{37} - \frac{35}{37} = -0.02702702702 \\
E(X^2) & = 17^2 \times p + q = \frac{17^2 \times 2}{37} + \frac{35}{37} = 16.5675675676 \\
Var(X) & = E(X^2) - E(X)^2 = 16.5675675676 - (-0.02702702702)^2 = 16.5668371074
\end{aligned}
$$

## Complex Strategies

### Black-Col-3 or Red-Col-2 Strategy

Betting on All blacks + betting on Column 3 (No.36)

- If $X$ lands on Column 3 and Red color: $X=1$ and $p_1=\frac{8}{37}$.
- If $X$ lands on Column 3 and Black color: $X=3$ and $p_2=\frac{4}{37}$
- If $X$ lands on Non-column 3 but Black: $X=0$ and $p_3=\frac{14}{37}$
- Otherwise you lose, with $X=-2$ and $q=1-p_1-p_2-p_3=\frac{11}{37}$

$$
\begin{aligned}
E(X) & = 1 \times p_1 + 3 \times p_2 + 1 * p_3 - 2 \times q = \frac{8}{37} + \frac{12}{37} - \frac{22}{37} = -\frac{2}{37} = -0.05405405405 \\
E(X^2) & = 1 \times p_1 + 9 \times p_2 + 4 \times q = \frac{8}{37} + \frac{36}{37} + \frac{44}{37} = \frac{88}{37} = 2.37837837838 \\
Var(X) & = E(X^2) - E(X)^2 = 2.37837837838 - 0.05405405405^2 = 2.37545653762
\end{aligned}
$$

Same strategy works when you bet on all Reds + betting on Column 2(No.35)

There's no need to calculate any other complex strategies. As any complex strategies can be regarded as an additive effects of simple strategies, therefore $E(X=\sum_{k=1}^nS_k)$.
