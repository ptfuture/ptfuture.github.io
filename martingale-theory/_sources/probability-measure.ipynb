# Probability Measure


## Probability space


$(\Omega,\mathcal{F},P)$ is called probability space. $\Omega$ is called **sample space**. $\mathcal{F}$ is a **$\sigma$-field**, which is a collection of sets of $\Omega$ such that:

1. if $A \in \mathcal{F}$, then $A^c \in \mathcal{F}$
2. if $A_i \in \mathcal{F}$ is countable sequence of sets then $\bigcup\limits_{i=1}^{\infty} A_i \in \mathcal{F}$

$P$ is a **measure** with $P(\Omega) = 1$. where a measure is a non-negative countable additive set function $\mu: \mathcal{F} \rightarrow \mathbf{R}$ with:

1. $\mu(A) \ge \mu(\emptyset) = 0$ for all $A \in \mathcal{F}$ and
2. If $A_i \in \mathcal{F}$ is a sequence of disjoint countable sets then, $\mu(\bigcup\limits_{i}^{\infty}A_i) = \sum\limits_{i}^{\infty} \mu(A_i)$

Properties of $P$:

1. Monotonicity: If $A \subset B$ then $P(B-A) = P(B)-P(A) \ge 0$
2. Subadditivity: If $A \subset \bigcup\limits_{n=1}^{\infty}A_n$ then $P(A) \le \sum\limits_{n=1}^{\infty}P(A_n)$.
3. Continuity from below: If $A_i \uparrow A$ then $P(A_i) \uparrow P(A)$.
4. Continuity from above: If $A_i \downarrow A$ then $P(A_i) \downarrow P(A)$.

**$\sigma(\mathcal{A})$** is called the smallest $\sigma$-field generated by $\mathcal{A}$. The existence of the smallest $\sigma$-field is because if $\mathcal{F_i}$ are all the $\sigma$-fields containing $\mathcal{A}$ then $\bigcap\limits_{i}\mathcal{F_i}$ is the smallest $\sigma$-field.


Let $\mathbf{R}$ be the real line, $\mathcal{R}$ the **Borel sets** which is the smallest $\sigma$-field containing the open sets. $\lambda$, the only measure on $\mathcal{R}$ with $\lambda((a,b])) = b-a$ for $a \lt b$ is called the **Lebesgue measure**. The existence and uniqueness of Lebesgue measure is proved in the **Caratheodory's Extension** Theorem.

Lebesgue measure to Probability measure: $\Omega = (0,1)$, $\mathcal{F} = \{A \cap (0,1): A \in \mathcal{R}\}$ and $P(A) = \lambda(A)$ for all $A \in \mathcal{F}$.

### Definition
A functino $X: \Omega \rightarrow S$ is said to be a **measurable map** from $(\Omega, \mathcal{F})$ to $(S, \mathcal{S})$ for every $B \in \mathcal{S}$ we have $X^{-1}(B) = \{\omega: X(\omega) \in B\} \in \mathcal{F}$. When $(S, \mathcal{S}) = (R^d, \mathcal{B^d})$, $X$ is called a **random vector** when $d \gt 1$ and **random variable** when $d = 1$. $X$ is also called **$\mathcal{F}$-measurable**.



### Theorem
Let $\mathcal{A}$ be a collection of subsets from a sample space $S$ with $\sigma(\mathcal{A}) = \mathcal{S}$, and let $X: \Omega \rightarrow S$ be a function. If $\{\omega: X(\omega) \in A\} \in \mathcal{F}$ for all $A \in \mathcal{A}$, then $X$ is a measurable map from $(\Omega, \mathcal{F})$ to $(S, \mathcal{S})$.



Proof: Let $\mathcal{S}^\circ = \{A: {X \in A} \in \mathcal{F}\}$, note that $\mathcal{S}^\circ \supset \mathcal{A}$. If $A_i \in \mathcal{S}^\circ$, $\bigcup_i\{X \in A_i\}=\{X \in \bigcup{A_i} \} \in \mathcal{F}$ indicates $\bigcup{A_i} \in \mathcal{S}^\circ$. Similarly, if $A \in \mathcal{S}^\circ$, $\{X \in A\}^c = \{X \in A^c\} \in \mathcal{F}$ indicates $A^c \in \mathcal{S}^\circ$. Therefore $\mathcal{S}^\circ$ is a $\sigma$-field. $\mathcal{S}^\circ \supset \sigma(\mathcal{A})$.

$\blacksquare$

It follows from the prove that, given measurable spaces $(\Omega, \mathcal{F})$, $(S, \mathcal{S})$ and $X: \Omega \rightarrow S$. The collection $\{ \{X \in B\}: B \in \mathcal{S} \}$ is a $\sigma$-field. It's called the $\sigma$-field generated by $X$.

### Definition
The probability measure $\mu$ induced by a random variable $X$ as $\mu(A) = P(X \in A)$ for Borel sets $A$ is called the **distribution** of $X$. $F(x) = P(X \le x)$ is called the **distribution function**.


<br />

Properties of distribution function:

1. Nondecreasing.
2. $\lim_{x \rightarrow\infty}F(x) = 1$, $\lim_{x \rightarrow -\infty}F(x) = 0$
3. Right continuous: $\lim_{y \downarrow x}F(y) = F(x)$
4. $F(x-) = \lim_{y \uparrow x} F(y) = P(X \lt x)$
5. $P(X = x) = F(x) - F(x-)$

Example of random variables:

- If $X: (\Omega, \mathcal{F}) \rightarrow (S, \mathcal{S})$, $f: (S, \mathcal{S}) \rightarrow (T, \mathcal{T})$ are measruable maps, then $f(X)$ is a measurable map from $(\Omega, \mathcal{F})$ to $(T, \mathcal{T})$. 
- If $X_1, X_2...X_n$ are r.v and $f: (\mathbf{R}^n, \mathcal{R}^n) \rightarrow (\mathbf{R}, \mathcal{R})$ is a measurable map, then $f(X_1,...,X_n)$ is a random variable.
- If $X_1, X_2...X_n$ are r.v, then $\inf\limits_{n}X_n$, $\sup\limits_{n}X_n$, $\limsup\limits_{n}X_n$ and $\liminf\limits_{n}X_n$ are all random variables.

$$
\begin{align}
\{ \inf\limits_{n}X_n \lt a\} = \bigcup_n\{ X_n \lt a\} \in \mathcal{F} \\
\{ \sup\limits_{n}X_n \lt a\} = \bigcap_n\{ X_n \lt a\} \in \mathcal{F}
\end{align}
$$

- $\lim_{n \rightarrow \infty}X_n$ exist iff the set $\{\limsup X_n \neq \liminf X_n\}$ has probabilty 0. Equivalently, we say $\limsup X_n = \liminf X_n$ almost surely/almost everywhere or in brief, a.s.

Limits on random variables may output infinity values, therefore, we need to extend our previous probability space to accomondate infinty values.

- $\mathbf{R}^{*} \equiv [-\infty, +\infty]$
- $\mathcal{R}^{*}$ Borel sets generated by intervals of the form $[-\infty, a), (a, b), (b, \infty]$

$(\mathbf{R}^{*}, \mathcal{R}^{*})$ forms a measurable space.

## Integration

**Jensen's Inequality** Suppose $\psi$ is convex, i.e. for $\lambda \in (0,1)$, $x,y \in \mathbf{R}$

$$
\begin{equation}
\lambda\psi(x) + (1-\lambda)\psi(y) \ge \psi({\lambda}x+(1-\lambda)y) (\#eq:convex-ineq)
\end{equation}
$$

Then:

$$
\begin{equation}
E(\psi(X)) \ge \psi(EX) (\#eq:jensen-ineq)
\end{equation}
$$

provided that both expectations exists. i.e. $E|X| \lt \infty$ and $E|\psi(X)| \lt \infty$.


**Holder's Inequality** Let $p,q \in [1,\infty]$ with $\frac{1}{p} + \frac{1}{q} = 1$ then:

$$
E|XY| \le \|X\|_p\|Y\|_q
$$
Where $|X\|_r=(E|X|^r)^{(1/r)}$ for $r \in [1,\infty)$; and $\|X\|_\infty = \inf\{M: P(|X| \gt M)=0\}$.

**Chebyshev's inequality** Suppose $\psi: \mathbf{R}\to\mathbf{R}$ and $\psi \ge 0$, Let $A \in \mathcal{R}$ and let $\alpha_A = \inf\{\psi(y): y \in A\}$:

$$
\begin{equation}
\alpha_AP(X \in A) \le E(\psi(X)\mathbf{1}_{(X \in A)}) \le E(\psi(X)) (\#eq:chebyshev-ineq)
\end{equation}
$$

_Proof:_ Take the integral on $\alpha_A\mathbf{1}_{(X \in A)} \le \psi(X)\mathbf{1}_{(X \in A)} \le \psi(X)$.

**Markov inequality:** Make $\psi(x) = x^2$, $A=\{x: |x| \ge a\}$, we get Markov inequality:

$$
\begin{equation}
a^2P(|X| \ge a) \le EX^2 (\#eq:markov-ineq)
\end{equation}
$$

**Monotone convergence theorem:** If $0 \le X_n \uparrow X$ a.s. then $EX_n \uparrow EX$.


**Fatou's lemma:** If $X_n \ge 0$, then:

$$
\begin{equation}
\liminf_{n\to\infty}EX_n \ge E(\liminf_{n\to\infty}X_n) (\#eq:fatou-ineq)
\end{equation}
$$

_Proof:_ If $Y_n = \inf_{k \ge n}X_k$, then $0 \le Y_n \uparrow Y = \liminf_{n\to\infty}X_n$, by the Monotone convergence theorem, ${\int}X_n{d\mu} \ge {\int}Y_n{d\mu} \uparrow {\int}Y{d\mu} = {\int}\liminf_{n\to\infty}X_n{d\mu}$.

$\blacksquare$




**Dominiated convergence theorem:** If $X_n \to X$ a.s., $|X_n| \le Y$ a.s. for all $n$ and $EY \lt \infty$, then $EX_n \to EX$.


**Product Measures:**

Note: If can be proved that if $\mu_1, \mu_2$ are two measures, then $\mu_1(A)\mu_2(B)$ is also a measure. We denote such measure as $\mu(A \times B)=\mu_1\times\mu_2(A \times B) = \mu_1(A)\mu_2(B)$ it's a unique measure on $\sigma(\mathcal{F_1}\times\mathcal{F_2})$(The smallest $\sigma$-field generated by $A\times B$ where $A\in\mathcal{F_1}, B\in\mathcal{F_2}$) such that $\mu(A \times B) = \mu_1(A)\mu_2(B)$

**Fubini Theorem: ** If $f \ge 0$ or $\int{|f|d\mu} \lt \infty$ then

$$
\int_{X{\times}Y}fd\mu = \int_{X}\int_{Y}f(x,y)\mu_1(dx)\mu_2(dy) = \int_{Y}\int_{X}f(x,y)\mu_2(dy)\mu_1(dx)
$$

**Corollary:** If $Y \ge 0$ and $p \gt 0$ then:

$$
\begin{equation}
E(Y^p) = \int_0^\infty{py^{p-1}P(Y \gt y)} dy
\end{equation}
$$

_Proof:_ Since $Y$ is positve, we can use Fubini theorem to exchange the integral:

$$
\begin{align}
\int_0^\infty{py^{p-1}P(Y \gt y)} dy &= \int_0^\infty \int_{\Omega}{py^{p-1}\mathbf{1}_{(Y \gt y)}}dP dy \\
&= \int_{\Omega}\int_0^Y {py^{p-1}}dydP \\
&= \int_{\Omega}Y^pdP = E(Y^p)
\end{align}
$$




### Theorem: Integrability implies finite almost surely

Let $(\Omega,\mathcal{F},\mu)$ be a measure space. If $f$ integrable w.r.t $\mu$, then $f$ is finite almost surely.

_Proof:_

$f$ is integrable implies: $\int_{\Omega}|f(x)| \mu(dx) \lt C$ for some constant $C$.

Definie $E_n=\{x: |f(x)| > n\}$. Then $E = \cap_n{E_n}$ is the set where $f$ goes to infinity. We want to prove $\mu(E) = 0$. Note that $n\mathbf{1}_{E_n} \le |f|$, we have:

$$
n\mu(E_n) \le \int_{E_n}{|f(x)|}\mu(dx) \le \int_{\Omega}{|f(x)|}\mu(dx) \lt C
$$

Since $E \subset E_n$ for all $n$, $0 \le \mu(E) \le \mu(E_n) \le \frac{C}{n}$ for all $n$. Let $n \to \infty$, we have $\mu(E) = 0$. i.e. $f$ is finite almost surely.

$\blacksquare$

## Sufficient and necessary conditions for convergence almost surely

#### Theorem 

If $\sum_{n=1}^{\infty} E[|X_n|^s] \lt \infty$ and $s \gt 0$, then $X_n \to 0$ a.s.


_Proof:_

This is an integration over positive function, by Fubini theorem, we can move expected value out of sumation: $E[\sum_{n=1}^{\infty} |X_n|^s] \lt \infty$. This implies $\sum_{n=1}^{\infty} |X_n|^s \lt \infty$ a.s. Further implies $|X_n|^s \to 0$ a.s., since $s \gt 0$, this implies $X_n \to 0$ a.s. $\blacksquare$

#### Theorem

$X_n \to 0$ a.s. iff for all $\epsilon \gt 0$, $P(|X_n| \gt \epsilon \text{ i.o}) = 0$

#### Theorem

If $\sum_{n=1}^{\infty}P(|X_n| \gt \epsilon) \lt \infty$ for all $\epsilon \gt 0$, then $X_n \to 0$ a.s. 

_Proof:_

Method 1: 

Fix $\epsilon = 1/k$ for any positive $k$, by the Borel-Cantelli lemma $\sum_{n=1}^{\infty}P(|X_n| \gt 1/k) \lt \infty$ implies $[|X_n| \gt 1/k \text{ i.o}]$ has probability 0. Equivalently, $[|X_n| \le 1/k \text{ e.a}]$ with probability one or, another equivalent interpretation is that $|X_n| \gt 1/k$ occurs finite number of times with probability one. Thus, $P(\limsup_{n\to\infty} |X_n| \gt 1/k) = 0$ for all positive $k$.

The set $\{ \limsup_{n\to\infty}|X_n| \gt 1/k \}$ enlarges as $k$ increases and converges to $\{ \limsup_{n\to\infty}|X_n| \gt 0 \}$. By the continuity from below property of probability measure, we have $0 = P(\{ \limsup_{n\to\infty}|X_n| \gt 1/k \}) \to P(\{ \limsup_{n\to\infty}|X_n| \gt 0 \})$ as $k\to\infty$. Therefore, $P(\{ \limsup_{n\to\infty}|X_n| \gt 0 \}) = 0$. And this impies $X_n \to 0$ a.s. $\blacksquare$

Method 2:

Let $\epsilon = 1/k$ for some positive $k$:

$\sum_{n=1}^{\infty}P(|X_n| \gt 1/k) = \sum_{n=1}^{\infty}E(\mathbf{1}_{(|X_n| \gt 1/k)}) \lt \infty$.

Since $\mathbf{1}_{(|X_n| \gt 1/k)}$ is positive, by the Fubini theorem, we could move the expection outside:

$$
E\Big(\sum_{n=1}^{\infty}\mathbf{1}_{(|X_n| \gt 1/k)}\Big) \lt \infty
$$

Which indicates $\sum_{n=1}^{\infty}\mathbf{1}_{(|X_n| \gt 1/k)} \lt \infty$ a.s., which implies $\mathbf{1}_{(|X_n| \gt 1/k)} \to 0$ a.s. as $n\to\infty$, or equivalently, $\{|X_n| \gt 1/k\}$ occurs finite number of times with probability one. And the rest of the prove will be the same as method 1. $\blacksquare$


#### Theorem

$X_n \to 0$ a.s. iff for every $\epsilon \gt 0$ we have $P(\sup_{m\ge n}|X_m| \gt \epsilon) \to 0$ as $n\to\infty$

_Proof:_

If an element $\omega$ such that $X_n\not\to 0$, it's equivalent to there exist a $\epsilon \gt 0$ such that $\limsup_{n\to\infty}|X_n|\gt \epsilon$. To translate it to set language we have:

$$
\{\omega: X_n \not\to 0\} \equiv \cup_{\epsilon \in \mathcal{Q^+}} \{\omega: \limsup_{n\to\infty} |X_n| \gt \epsilon\} \equiv \cup_{\epsilon \in \mathcal{Q^+}} \cap_{n} \{\omega: \sup_{m \ge n} |X_m| \gt \epsilon \}
$$

$A_{\epsilon,n} = \{\omega: \sup_{m \ge n} |X_m| \gt \epsilon \}$ decreases as $n\to \infty$ and $A_{\epsilon,n} \downarrow \cap_{n} A_{\epsilon,n}$. By the probability measure's the continuity from above property, we have $P(A_{\epsilon,n})\downarrow P(\cap_{n} A_{\epsilon,n})$.

$\Rightarrow:$ Given $P(X_n\not\to 0) = 0$, since $\{\omega: X_n \not\to 0\} \supset \cap_{n} A_{\epsilon,n}$ for any $\epsilon \in \mathcal{Q^+}$, we get $P(\cap_{n} A_{\epsilon,n}) = 0$ and $P(A_{\epsilon,n})\to P(\cap_{n} A_{\epsilon,n}) = 0$ for all $\epsilon \gt 0$.

$\Leftarrow:$ Given for every $\epsilon \gt 0$, $P(A_{\epsilon,n}) \to 0 = P(\cap_{n} A_{\epsilon,n})$ as $n\to\infty$. We have: $P(X_n\not\to 0) \le \sum_{\epsilon\in\mathcal{Q^+}}P(\cap_{n} A_{\epsilon,n}) \to 0$ as $n\to\infty$.

$\blacksquare$
